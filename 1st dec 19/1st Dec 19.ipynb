```python
students = pd.read_excel('./students.xlsx')
students_contacts = pd.read_excel('./students_contacts.xlsx')
```

```python
pd.merge(students, 
         students_contacts, 
         left_index=True,
         right_on='Roll Number',
         how='outer')
```
